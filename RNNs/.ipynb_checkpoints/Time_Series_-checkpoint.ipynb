# Train - test - split

```python
len(df)
```

```python
test_size = 18

test_ind = len(df)-test_size
```

```python
train = df.iloc[:test_ind]
test = df.iloc[test_ind:]
```

```python
train.shape,test.shape
```

# Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scale_train = scaler.fit_transform(train)
scale_test = scaler.transform(test)
```

# TimeseriesGenerator

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
len(test)
```

```python
length = 12

generator = TimeseriesGenerator(scale_train,scale_train,
                                length=length,batch_size=1)
```

```python
x,y = generator[0]
```

```python
x
```

```python
y
```

# LSTM Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,LSTM
from tensorflow.keras.callbacks import EarlyStopping
```

```python
n_feature = 1
```

```python
model = Sequential()
model.add(LSTM(100,activation='relu',input_shape=(length,n_feature)))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')

model.summary()
```

```python
early_stop = EarlyStopping(patience=2)
```

```python
validation_generator = TimeseriesGenerator(scale_test,scale_test,
                                length=length,batch_size=1)
```

```python
model.fit_generator(generator,epochs=20,
                    validation_data=validation_generator,
                    callbacks=[early_stop])
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss.plot(figsize=(12,8))
```

# prediction main logic

```python
test_predictions = []

first_eval_batch = scale_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_feature))


for i in range(len(test)):
    
    current_pred = model.predict(current_batch)[0]
    
    test_predictions. append(current_pred)
    
    current_batch = np.append(current_batch[:,1:,],[[current_pred]],axis=1)
```

```python
prediction = scaler.inverse_transform(test_predictions)
```

```python
test['prediction'] = prediction
```

```python
test
```

```python
test.plot(figsize=(12,8))
```

# Training on Full DAta 

```python
full_scale = scaler.fit_transform(df)
```

```python
length=12

generator = TimeseriesGenerator(full_scale,full_scale,length=length,batch_size=1)
```

```python
model = Sequential()
model.add(LSTM(100,activation='relu',input_shape=(length,n_feature)))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')

model.summary()
```

```python
model.fit_generator(generator,epochs=5)
```

# Forecast Logic

```python
forecast = []

period = 12    
    
first_eval_batch = full_scale[-length:]
current_batch = first_eval_batch.reshape((1,length,n_feature))


for i in range(period):
    
    current_pred = model.predict(current_batch)[0]
    
    forecast.append(current_pred)
    
    current_batch = np.append(current_batch[:,1:,],[[current_pred]],axis=1)
```

```python
forecast = scaler.inverse_transform(forecast)
```
